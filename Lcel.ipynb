```python
import langchain

import os
from dotenv import load_dotenv
load_dotenv()

groq_api_key = os.getenv("groq_api_key")
groq_api_key
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(groq_api_key=groq_api_key, model="llama-3.1-8b-instant")
model
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

message =[
    SystemMessage(content="""
                  Translate the following sentence into Spanish.
Return ONLY the translated sentence.
Do not include any explanation or reasoning"""),
    HumanMessage(content="hello! what are you doing?")
]
result = model.invoke(message)
```

```python
print(result)
```

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()

parser.invoke(result)
```

```python
### Using LCEL- chain the components
chain = model|parser

chain.invoke(message)
```

```python
### Prompt Templates
from langchain_core.prompts import ChatPromptTemplate
genral_prompt = """Translate the following sentence into Spanish.
Return ONLY the translated sentence.
Do not include any explanation or reasoning"""
prompt = ChatPromptTemplate.from_messages([
    ("system", genral_prompt),
    ("user", "{input}")
])
```

```python
chain = prompt|model|parser

chain.invoke({"input": "hello! what are you doing?"})

```
